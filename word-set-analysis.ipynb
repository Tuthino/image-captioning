# Word Number Analysis

Let's compare the amount of words in the training set and test set, and see which have words in common.

```python

def update_set(this_set, text):
    
    if type(text) == float: return this_set
    
    new_words = set(text.split(' '))
    #
    new_words = set(re.sub(r'[^a-zA-Z0-9\-]', '', word) for word in new_words)
    
    if new_words: return this_set.update(new_words)
    
    return this_set

words_train = set({})
for text, _ in train_set:
    update_set(words_train, text)

words_test = set({})
for text, _ in test_set: update_set(words_test, text)
for text, _ in val_set: update_set(words_test, text)

print("Number of Training Examples:", len(train_set), "\n")

print("Num of train set words" , len(words_train))
print("Num of test set words" , len(words_test))

print("Percentage of common words, that occur in test set:")
print(len(words_train.intersection(words_test)) / len(words_test))
```

Even though the train set has ~6000 words, more than the set set, which has ~2700 words, 30 % of test set words are seen during training.

```python
from tokenizer import *


def update_set(this_set, text):
    
    if type(text) == float: return this_set
    
    token_data = tokenizer.encode(text)
    tokens = set(number for number in token_data)
    
    if tokens: return this_set.update(tokens)
    
    return this_set

tokens_train = set({})
for text, _ in train_set:
    update_set(tokens_train, text)

tokens_test = set({})
for text, _ in test_set: update_set(tokens_test, text)
for text, _ in val_set: update_set(tokens_test, text)

print("Number of Training Examples:", len(train_set), "\n")

print("Num of unique train tokens" , len(tokens_train))
print("Num of unique test tokens" , len(tokens_test))

print("Percentage of test tokens, that occur in train set:")
print(len(tokens_train.intersection(tokens_test)) / len(tokens_test))
```
